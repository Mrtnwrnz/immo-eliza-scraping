```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
url = "https://www.immoweb.be/nl/zoekertje/uitzonderlijk-vastgoed/te-koop/schoten/2900/20228162"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html")

data_dict = {}
url_split = url.split("/")

# add property type
# if url from list houses:      (ALSO POSSIBLE: if subtype is in list of subtypes for house)
data_dict["property_type"] = "house"

# add price
price = soup.find("p", attrs={"class": "classified__price"}).text.split(" ")[-1].strip()
if price[-1] == "€": price = price[:-1]
data_dict["price"] = price

# data from url
data_dict["property_id"] = url_split[-1]
data_dict["postal_code"] = url_split[-2]
data_dict["locality_name"] = url_split[-3]
data_dict["property_subtype"] = url_split[-5]

# add all data from classified table
rows = soup.find_all('tr', class_='classified-table__row')
for row in rows:
    header = row.find('th', class_='classified-table__header')
    data_cell = row.find('td', class_='classified-table__data')
    if header and data_cell:
        header_key = header.contents[0].strip()
        data_cell_value = data_cell.contents[0].strip()
    else:
        print(f"error for {url}: header = {header}, data_cell = {data_cell}")
    data_dict[header_key] = data_cell_value

# check for life sales
if 'Maandelijkse rente' in data_dict:
    print('yes')
else:
    print('no')

# add number of rooms
# more types of rooms?
rooms = 0
if 'Oppervlakte woonkamer' in data_dict.keys():
    rooms +=1
if 'Oppervlakte keuken' in data_dict.keys():
    rooms +=1
if 'Slaapkamers' in data_dict.keys():
    rooms += int(data_dict['Slaapkamers'])
if 'Badkamers' in data_dict.keys():
    rooms += int(data_dict['Badkamers'])
if 'Kelder' in data_dict.keys():
    rooms +=1
data_dict['rooms'] = rooms

# add type sale


#  add missing mandatory values
columns = ['property_type', 'price', 'property_id', 'postal_code', 'locality_name', 'property_subtype', 'Staat van het gebouw', 'Aantal gevels', 'Bewoonbare oppervlakte', 'Type keuken', 'Oppervlakte tuin', 'Oppervlakte terras', 'Gemeubeld', 'Aantal open haarden', 'Zwembad', 'rooms', 'sale_type']
for i in columns:
    if i in data_dict.keys():
        pass
    else:
        data_dict[i] = "None"

link_1 = "https://www.immoweb.be/nl/zoekertje/huis/te-koop/bonheiden/2820/20226641"
link_open_haard = "https://www.immoweb.be/nl/zoekertje/huis/te-koop/uccle/1180/20226959"
link_zwembad = "https://www.immoweb.be/nl/zoekertje/uitzonderlijk-vastgoed/te-koop/schoten/2900/20228162"
link_gemeubeld = "https://www.immoweb.be/nl/zoekertje/huis/te-koop/oudenaarde/9700/20229990"
link_lijfrente = "https://www.immoweb.be/nl/zoekertje/huis/te-koop/amay/4540/20215296"
print(data_dict)



```
